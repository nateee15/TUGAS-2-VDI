## **2. Visualisasi Data yang Kurang Baik**

image-2.png

sumber : https://geospasialis.com/visualisasi-data/

* **Kejelasan (Clarity):**

  Visualisasi ini cukup jelas dengan garis yang menghubungkan data pada grafik. Titik-titik data yang terhubung memudahkan pembaca untuk melihat pola dalam data deret waktu (time series). Judulnya menjelaskan secara spesifik bahwa ini adalah scatterplot terhubung untuk data deret waktu.
Namun label sumbu Y tidak mencakup keterangan yang menjelaskan apa yang diukur, sehingga mungkin membingungkan pengguna baru.

* **Akurasi (Accuracy)**

   Grafik ini menggunakan skala yang sesuai untuk menampilkan data, dengan data yang terhubung secara akurat dari waktu ke waktu. Namun tidak ada keterangan jelas mengenai unit pengukuran pada sumbu Y. Meskipun pola data terlihat, kurangnya informasi detail bisa mempengaruhi interpretasi pengguna.

* **Efisiensi (Efficiency):**

  Informasi tentang tren waktu dapat ditangkap dengan cepat hanya dengan melihat garis yang menghubungkan titik-titik data. Visualisasi ini memudahkan pengguna untuk memahami pola data dalam rentang waktu tertentu. Namun bebrapa pengguna mungkin perlu menelusuri lebih jauh untuk mendapatkan rincian atau penjelasan mengenai nilai pada sumbu Y dan apa yang diukur.

* **Daya Tarik (Engagement):**

   Warna grafik bisa dibuat sedikit lebih kontras agar lebih menonjol. Selain itu, visualisasinya tidak terlalu mencolok, sehingga bisa kurang menarik bagi audiens yang lebih visual.

* **Kejelasan (Clarity):**

   Peta dunia ini menunjukkan distribusi bahasa hidup di seluruh dunia dengan jelas menggunakan titik-titik kecil di lokasi asal bahasa tersebut. Judulnya singkat dan tepat, memberikan konteks yang langsung kepada audien. Namun tidak ada legenda atau keterangan yang menjelaskan apa yang dimaksud dengan titik-titik kecil atau apakah ada variasi ukuran atau warna untuk mewakili sesuatu (misalnya jumlah penutur). Hal ini bisa membuat beberapa informasi kurang jelas.

* **Akurasi (Accuracy)**

  Visualisasi ini tampaknya akurat dalam hal lokasi geografis asal bahasa-bahasa yang ditunjukkan. Data ditempatkan di lokasi yang tepat di peta dunia. Namun, tidak ada informasi lebih lanjut mengenai skala atau representasi kuantitatif dari titik-titik tersebut. Apakah setiap titik mewakili satu bahasa atau lebih? Ini bisa mempengaruhi pemahaman pengguna tentang distribusi sebenarnya.

* **Efisiensi (Efficiency):**

 Efisiensi dapat ditingkatkan dengan menambahkan lebih banyak detail, seperti kategori atau warna yang menunjukkan jumlah penutur atau klasifikasi bahasa berdasarkan kelompok tertentu.

* **Daya Tarik (Engagement):**

  Kurangnya variasi warna atau simbol mungkin membuat visualisasi ini sedikit monoton dan kurang interaktif. Menambahkan elemen visual yang lebih dinamis atau interaktif bisa meningkatkan keterlibatan pengguna.
